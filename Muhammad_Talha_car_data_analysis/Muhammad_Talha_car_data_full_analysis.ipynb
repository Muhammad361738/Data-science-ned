```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./car_data.csv')
```

```python
df.head(1)
```

# Data Analysis Methods for Car Dataset

# 1. **Highest Price**: Finds the highest price in the dataset.


```python
highest_price = df['price'].max()
print(highest_price)
```

## 2. **Lowest Price**: Finds the lowest price in the dataset.


```python
lowest_price = df['price'].min()
print(lowest_price)
```

### 3. **Average Price**: Calculates the average price of all cars.


```python
average_price = df['price'].mean()
print(average_price)
```

### 4. **Car with Highest Horsepower**: Finds the car with the highest horsepower.


```python
car_highest_hp = df.loc[df['horsepower'].idxmax()]
print(car_highest_hp)
```

### 5. **Car with Lowest Horsepower**: Finds the car with the lowest horsepower.


```python
car_lowest_hp = df.loc[df['horsepower'].idxmin()]
print(car_lowest_hp)
```

### 6. **Average Horsepower**: Calculates the average horsepower across all cars.


```python
average_hp = df['horsepower'].mean()
print(average_hp)
```

### 7. **Number of Unique `carbody` Types**: Counts the number of unique `carbody` types in the dataset.


```python
unique_carbody_types = df['carbody'].nunique()
print(unique_carbody_types)
```

### 8. **Number of Cars by `fueltype`**: Counts how many cars there are for each fuel type (e.g., gas, diesel).


```python
cars_by_fueltype = df['fueltype'].value_counts()
print(cars_by_fueltype)
```

### 9. **Median Price**: Finds the median value of `price` (middle value when sorted).


```python
median_price = df['price'].median()
print(median_price)
```

### 10. **Standard Deviation of Price**: Measures the spread of `price` values.


```python
price_std_dev = df['price'].std()
print(price_std_dev)
```

### 11. **Correlation Between `price` and `horsepower`**: Shows the correlation between price and horsepower (how closely they are related).


```python
price_hp_correlation = df[['price', 'horsepower']].corr().iloc[0, 1]
print(price_hp_correlation)
```

### 12. **Count of Cars by `carbody`**: Counts how many cars there are of each `carbody` type.


```python
count_by_carbody = df['carbody'].value_counts()
print(count_by_carbody)
```

### 13. **Sum of All Engine Sizes**: Calculates the sum of the `enginesize` column.


```python
sum_enginesize = df['enginesize'].sum()
print(sum_enginesize)
```

### 14. **Unique Values in `aspiration`**: Finds the unique values in the `aspiration` column (e.g., std, turbo).


```python
unique_aspiration = df['aspiration'].unique()
print(unique_aspiration)
```

### 15. **Count of Cars with Turbo Aspiration**: Counts how many cars have a `turbo` aspiration.


```python
count_turbo = df[df['aspiration'] == 'turbo'].shape[0]
print(count_turbo)
```

### 16. **Car with Highest Engine Size**: Finds the car with the highest engine size.


```python
car_highest_enginesize = df.loc[df['enginesize'].idxmax()]
print(car_highest_enginesize)
```

### 17. **Car with Lowest Engine Size**: Finds the car with the lowest engine size.


```python
car_lowest_enginesize = df.loc[df['enginesize'].idxmin()]
print(car_lowest_enginesize)
```

### 18. **Number of Cars with High Mileage**: Counts the number of cars with high `citympg` (e.g., more than 30 miles per gallon).


```python
high_mileage_cars = df[df['citympg'] > 30].shape[0]
print(high_mileage_cars)
```

### 19. **Car with Maximum Wheelbase**: Finds the car with the maximum `wheelbase`.


```python
car_max_wheelbase = df.loc[df['wheelbase'].idxmax()]
print(car_max_wheelbase)
```

### 20. **Range of `price` (Max - Min)**: Calculates the range of car prices (i.e., difference between max and min price).


```python
price_range = df['price'].max() - df['price'].min()
print(price_range)
```

### 21. **Count of Cars by `drivewheel` Type**: Counts how many cars there are of each `drivewheel` type (e.g., fwd, rwd).


```python
count_by_drivewheel = df['drivewheel'].value_counts()
print(count_by_drivewheel)
```

### 22. **Top 5 Expensive Cars**: Lists the top 5 cars with the highest prices.


```python
top_5_expensive_cars = df.nlargest(5, 'price')[['CarName', 'price']]
print(top_5_expensive_cars)
```

### 23. **Top 5 Cheapest Cars**: Lists the top 5 cars with the lowest prices.


```python
top_5_cheapest_cars = df.nsmallest(5, 'price')[['CarName', 'price']]
print(top_5_cheapest_cars)
```

### 24. **Count of Cars with Missing Values**: Counts how many rows have missing values in any column.


```python
missing_values = df.isnull().sum().sum()
print(missing_values)
```

# Data Visualization Methods for Car Dataset

### 1. **Box Plot for Price by `carbody`**:



```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='carbody', y='price', data=df)
plt.title('Price Distribution by Car Body Type')
plt.show()
```

### 2. **Histogram for `horsepower` Distribution**:


```python
plt.figure(figsize=(10, 6))
sns.histplot(df['horsepower'], bins=20, kde=True)
plt.title('Horsepower Distribution')
plt.show()
```

### 3. **Scatter Plot for `horsepower` vs `price`**:


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='horsepower', y='price', data=df)
plt.title('Horsepower vs Price')
plt.show()
```

### 4 **Pairplot for Multiple Variables**:


```python
sns.pairplot(df[['price', 'horsepower', 'enginesize', 'curbweight']])
plt.title('Pairplot of Price, Horsepower, Engine Size, and Curb Weight')
plt.show()
```

### 5. **Correlation Heatmap for Numerical Columns**:


```python
plt.figure(figsize=(10, 8))
sns.heatmap(df[['price', 'horsepower', 'enginesize', 'curbweight']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

### 6 . **Bar Chart for Number of Cars by `fueltype`**:


```python
plt.figure(figsize=(8, 6))
sns.countplot(x='fueltype', data=df)
plt.title('Number of Cars by Fuel Type')
plt.show()
```

### 7 . **Count Plot for `carbody`**:


```python
plt.figure(figsize=(8, 6))
sns.countplot(x='carbody', data=df)
plt.title('Number of Cars by Car Body Type')
plt.show()
```

### 8 **Scatter Plot for `curbweight` vs `enginesize`**:


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='curbweight', y='enginesize', data=df)
plt.title('Curb Weight vs Engine Size')
plt.show()
```
